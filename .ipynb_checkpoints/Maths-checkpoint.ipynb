```python
import numpy as np
import matplotlib.pyplot as mpl
```

```python
x = np.linspace(-1,50,500)
```

```python
mpl.plot(x,2**x,label='2**x')
mpl.plot(x,(np.log(2)*(2**x)),label='log(2)*(2**x)')
mpl.grid(1)
mpl.legend()
mpl.show()
```

```python
np.arcsin(0.5)
```

```python
mpl.plot(x,360*(np.arcsin(x)/(2*np.pi)))
```
