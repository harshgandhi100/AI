```python
import numpy as np
import matplotlib.pyplot as mpl
```

```python
x = np.linspace(-10,10,500)
```

```python
mpl.plot(x,1/(1+np.exp(-x)),label='Sigmoid')
mpl.plot(x,np.exp(-x)/((1+np.exp(-x))**2),label='Sigmoid Prime')
mpl.grid(1)
mpl.legend()
mpl.show()
```
